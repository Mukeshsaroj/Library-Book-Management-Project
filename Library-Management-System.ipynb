# Liabrary Management System

```python
from tkinter import *
import tkinter as tk
import sqlite3
import tkinter.ttk as ttk
import tkinter.messagebox as tkMassgeBox
```

```python
fullframe=Tk()
fullframe.title('Library Managment System')

def database():
    global conn,cursorobj
    conn=sqlite3.connect('books.db')
    cursorobj=conn.cursor()
    cursorobj.execute("CREATE TABLE IF NOT EXISTS librarytable(authorname TEXT,bookname TEXT,publishedyear INTEGER,isbn INTEGER)")
    

def displaybook():
    tree.delete(*tree.get_children())
    database()
    cursorobj.execute("SELECT * FROM librarytable ORDER BY bookname ASC")
    fetch=cursorobj.fetchall()
    for data in fetch:
        tree.insert('','end',values=(data[0],data[1],data[2],data[3]))
    cursorobj.close()
    conn.close()
    tk.messagebox.showinfo('Liabrary','Data fetched')
    
def searchbook():
    titleauthor=((authorname.get()).title())
    titlebookname=((bookname.get()).title())
    titleisbn=((isbn.get()).title())
    tree.delete(*tree.get_children())
    database()
    cursorobj.execute("SELECT * FROM librarytable")
    fetch=cursorobj.fetchall()
    count=0
    for data in fetch:
        if(titleauthor==data[0]) and (titlebookname==data[1]) and (str(publishedyear.get())==str(data[2])) and (str(titleisbn)==str(data[3])):
            tree.insert('','end',values=(data[0],data[1],data[2],data[3]))
            count+=1
        elif(titleauthor==data[0]):
            tree.insert('','end',values=(data[0],data[1],data[2],data[3]))
            count+=1
        elif(titlebookname==data[1]):
            tree.insert('','end',values=(data[0],data[1],data[2],data[3]))
            count+=1
        elif(str(publishedyear.get())==str(data[2])):
            tree.insert('','end',values=(data[0],data[1],data[2],data[3]))
            count+=1
        elif(str(titleisbn)==str(data[3])):
            tree.insert('','end',values=(data[0],data[1],data[2],data[3]))
            count+=1
    if(count==0):
        tk.messagebox.showinfo('Library','No book like this in library')
    cursorobj.close()
    conn.close()
    authorname.set("")
    bookname.set("")
    publishedyear.set("")
    isbn.set("")
    
    
    
def addbook():
    titleauthor=((authorname.get()).title())
    titlebookname=((bookname.get()).title())
    titleisbn=((isbn.get()).title())
    if ((authorname.get()=='') or (bookname.get()=='') or (publishedyear.get()=='') or (isbn.get()=='')):
        tk.messagebox.showinfo('Library','Please fill all the details')
    else:
        database()
        cursorobj.execute("INSERT INTO librarytable(authorname,bookname,publishedyear,isbn) VALUES(?,?,?,?)",(titleauthor,titlebookname,publishedyear.get(),titleisbn))
        conn.commit()
        authorname.set("")
        bookname.set("")
        publishedyear.set("")
        isbn.set("")
        cursorobj.close()
        conn.close()
        tk.messagebox.showinfo('Library','Values inserted')
        
def deletebook():
    titleauthor=((authorname.get()).title())
    titlebookname=((bookname.get()).title())
    titleisbn=((isbn.get()).title())
    titlepublishedyear=(publishedyear.get().title())
    tree.delete(*tree.get_children())
    database()
    cursorobj.execute("SELECT * FROM librarytable")
    fetch=cursorobj.fetchall()
    count=0
    for data in fetch:
        if ((titleauthor==data[0]) and (titlebookname==data[1]) and (str(titlepublishedyear)==str(data[2])) and (str(titleisbn)==str(data[3]))):
            tempa=("'"+str(titleauthor)+"'")
            tempb=("'"+str(titlebookname)+"'")
            tempp=("'"+str(titlepublishedyear)+"'")
            tempi=("'"+str(titleisbn)+"'")
            temp1="DELETE FROM librarytable WHERE authorname= "+tempa +"and bookname= "+tempb +"and publishedyear= "+tempp +"and isbn= "+tempi
            cursorobj.execute(temp1)
            conn.commit()
            count+=1
            
    if(count==0):
        tk.messagebox.showinfo('Library','No data with this name')
    else:
        tk.messagebox.showinfo('Library','Book deleted')
    authorname.set("")
    bookname.set("")
    publishedyear.set("")
    isbn.set("")
    conn.commit()
    cursorobj.close()
    conn.close()
    
    
def issuebook():
    titleauthor=((authorname.get()).title())
    titlebookname=((bookname.get()).title())
    titleisbn=((isbn.get()).title())
    titlepublishedyear=(publishedyear.get().title())
    tree.delete(*tree.get_children())
    database()
    cursorobj.execute("SELECT * FROM librarytable")
    fetch=cursorobj.fetchall()
    cursorobj.execute("DELETE FROM librarytable")
    count=0
    for data in fetch:
        if(titleauthor==data[0]) and (titlebookname==data[1]) and (titlepublishedyear==str(data[2])) and (titleisbn==str(data[3])):
            if(count==0):
                count+=1
                tk.messagebox.showinfo('Library','Book issued')
            else:
                cursorobj.execute("INSERT INTO librarytable(authorname,bookname,publishedyear,isbn) VALUES(?,?,?,?)",(data[0],data[1],data[2],data[3]))
                conn.commit()
        else:
            cursorobj.execute("INSERT INTO librarytable(authorname,bookname,publishedyear,isbn) VALUES(?,?,?,?)",(data[0],data[1],data[2],data[3]))
            conn.commit()
    if(count==0):
        tk.messagebox.showinfo('Library','No book with this detail in library')
    authorname.set("")
    bookname.set("")
    publishedyear.set("")
    isbn.set("")
    conn.commit()
    cursorobj.close()
    conn.close()
        
    
    
    
    
def Exit():
    result=tkMassgeBox.askquestion('Library','Do you want to exit?',icon='warning')
    if(result=='yes'):
        fullframe.destroy()
    else:
        tk.messagebox.showinfo('Library','You will now return to the application screen')    


#=============================Frames==================================================================
TitleFrame=Frame(fullframe,width=1350,padx=20,bd=20,relief=RIDGE,bg='yellow')
TitleFrame.pack(side=TOP)
lblTitle=Label(TitleFrame,width=45,font=('Lucida Sans',40,'bold'),text='Library Managment System!!!',padx=12,bg='yellow', fg='#f00')
lblTitle.grid()



ButtonFrame=Frame(fullframe,bd=20,padx=20,relief=RIDGE,bg='yellow')
ButtonFrame.pack(side=BOTTOM)

LeftFrame=Frame(fullframe,bd=20,width=620,height=520,padx=20,relief=RIDGE,bg='yellow')
LeftFrame.pack(side=LEFT)

Form=Frame(LeftFrame,bd=20,width=620,height=520,padx=20,bg='yellow')
Form.pack()

RightFrame=Frame(fullframe,bd=20,width=930,height=520,padx=20,relief=RIDGE,bg='yellow')
RightFrame.pack(side=RIGHT)

#===========================================Variables===========================================

authorname=StringVar()
bookname=StringVar()
publishedyear=StringVar()
isbn=StringVar()

#===========================================Label&Entry===========================================

lbl_authorname=Label(Form,text="Author Name:",font=('Rockwell',18,'bold'),bg='yellow', fg='#f00')
lbl_authorname.grid(row=0,column=0,sticky='w')
entry_authorname=Entry(Form,textvariable=authorname,width=35)
entry_authorname.grid(row=0,column=1)

lbl_bookname=Label(Form,text="Book Name:",font=('Rockwell',18,'bold'),bg='yellow', fg='#f00')
lbl_bookname.grid(row=1,column=0,sticky='w')
entry_bookname=Entry(Form,textvariable=bookname,width=35)
entry_bookname.grid(row=1,column=1)

lbl_publishyear=Label(Form,text="Published year:",font=('Rockwell',18,'bold'),bg='yellow', fg='#f00')
lbl_publishyear.grid(row=2,column=0,sticky='w')
entry_publishyear=Entry(Form,textvariable=publishedyear,width=35)
entry_publishyear.grid(row=2,column=1)

lbl_ISBN=Label(Form,text="ISBN No:",font=('Rockwell',18,'bold'),bg='yellow', fg='#f00')
lbl_ISBN.grid(row=3,column=0,sticky='w')
entry_ISBN=Entry(Form,textvariable=isbn,width=35)
entry_ISBN.grid(row=3,column=1)


#===========================================Buttons===========================================

but_view=Button(ButtonFrame,text="View Library",width=19,font=('Rockwell',15,'bold'),command=displaybook,bg='yellow', fg='#f00')
but_view.grid(row=0,column=0)

but_search=Button(ButtonFrame,text="Search",width=19,font=('Rockwell',15,'bold'),command=searchbook,bg='yellow', fg='#f00')
but_search.grid(row=0,column=1)

but_add=Button(ButtonFrame,text="Add Book",width=19,font=('Rockwell',15,'bold'),command=addbook,bg='yellow', fg='#f00')
but_add.grid(row=0,column=2)

but_issue=Button(ButtonFrame,text="Issue Book",width=19,font=('Rockwell',15,'bold'),command=issuebook,bg='yellow', fg='#f00')
but_issue.grid(row=1,column=0)

but_delete=Button(ButtonFrame,text="Delete Book",width=19,font=('Rockwell',15,'bold'),command=deletebook,bg='yellow', fg='#f00')
but_delete.grid(row=1,column=1)

but_exit=Button(ButtonFrame,text="Exit",width=19,font=('Rockwell',15,'bold'),command=Exit,bg='yellow', fg='#f00')
but_exit.grid(row=1,column=2)



#===========================================Scrollbar===========================================

scrollbary=Scrollbar(RightFrame)
scrollbarx=Scrollbar(RightFrame)

col=('Author_Name','Book_Name','Published_year','ISBN')
tree=ttk.Treeview(RightFrame,columns=col,show='headings')
scrollbary=Scrollbar(RightFrame)
scrollbary.pack(side=RIGHT,fill=Y)
scrollbarx=Scrollbar(RightFrame)
scrollbarx.pack(side=BOTTOM,fill=X)


tree.heading("Author_Name",text="Author Name",anchor=W)
tree.heading("Book_Name",text="Book Name",anchor=W)
tree.heading("Published_year",text="Published year",anchor=W)
tree.heading("ISBN",text="ISBN",anchor=W)

tree.column("Author_Name",minwidth=0,width=250)
tree.column("Book_Name",minwidth=0,width=250)
tree.column("Published_year",minwidth=0,width=300)
tree.column("ISBN",minwidth=0,width=250)
tree.pack(side=TOP,fill=BOTH)

fullframe.configure(bg='yellow')
fullframe.mainloop()
```
